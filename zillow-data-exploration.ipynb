<b>Here we see the declining trend in New Jersey's Unemployment Rate</b><sup>1</sup>, frome 2012-07 to 2013-04 when unemployment declined by 0.1% each month.  A preliminary hypothesis is that although it is illegal is some (to my knowledge, if not all) places to deny tenants who are unemployed, I expect there to be at least minimal correlation between levels in renting and unemployment rate.  Consequently, as demand for rentals may increase as unemployment decreases, it can be expected to find a slightly upward trend in median rent prices between 2012-07 and 2013-04.

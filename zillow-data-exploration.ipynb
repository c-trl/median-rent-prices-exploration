<b>Objective</b>: Perform a basic statistical analysis using Zillow Rent Index (ZRI)

<i>For this exercise, pandas will be used to handle much of the data manipulation, cleaning, and preparation for visualization with matplotlib.</i>

```python
#setup
# -*- coding: utf-8 -*-
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

---

<b>Preliminary data restructuring...</b>

<i>Raw data (uploaded to github) is assigned to a variable psqftZRI (for per square foot ZRI - no, I did not just mash some keys on my keyboard), so it can be called up easily later.</i>

```python
psqftZRI = 'https://raw.githubusercontent.com/c-trl/median-rent-prices-exploration/master/City_ZriPerSqft_AllHomes.csv'
dpsf = pd.read_csv(psqftZRI)
```

<i>A significant amount of data is missing for the dataset, so interpolation is performed to populate gaps in data.  Extraneous data is also dropped from the dataset, which is then written to a new csv file for a quick transpose - pandas wouldn't transpose the way I wanted it to.</i>

```python
#re-organizing ZRI data
df = dpsf.sort(['State', 'RegionName']).interpolate()
df = df.drop('Metro',1).drop('CountyName',1)
df = df.reset_index(level=1).drop('index',1)
df.rename(columns={'RegionName':'Region'}, inplace=True)
df['Location'] = df.Region.map(str) + ", " + df.State
df = df.drop('Region',1,).drop('State',1)

#moving Location to 0th position in the dataframe
cols = df.columns.tolist()
cols = cols[-1:] + cols[:-1]
df = df[cols]

#write resulting data to a csv file
#df.to_csv('df.csv')
```

<i>After going through the new csv file, I realized that if I wanted to compare places like Jersey City, NJ to Lower East Side in Manhattan, I would have to pull in neighborhood-level data from Zillow and merge it with city-level data.  The code used to clean the first dataset is used against neighborhood-level NYC data here:</i>

```python
nyc = pd.read_csv('https://raw.githubusercontent.com/c-trl/median-rent-prices-exploration/master/nyc.csv')

#cleaning NYC data
nydf = nyc.sort(['State', 'RegionName']).interpolate()
nydf = nydf.reset_index(level=1).drop('index',1)
nydf.rename(columns={'RegionName':'Region'}, inplace=True)
nydf['Location'] = nydf.Region.map(str) + ", " + nydf.State
nydf = nydf.drop('Region',1,).drop('State',1)

#moving Location to 0th position in the dataframe
cols = nydf.columns.tolist()
cols = cols[-1:] + cols[:-1]
nydf = nydf[cols]

#write resulting data to a csv file for further cleaning (transpose, remove spaces + commas)
#nydf.to_csv('nydf.csv')

#Appending New York City data
nyc = pd.read_csv('https://raw.githubusercontent.com/c-trl/median-rent-prices-exploration/master/nydf.csv')
```

<b>Accessing modified data...</b>

```python
#open cleaned csv
df = pd.read_csv('https://raw.githubusercontent.com/c-trl/median-rent-prices-exploration/master/df.csv')
nydf = pd.read_csv('https://raw.githubusercontent.com/c-trl/median-rent-prices-exploration/master/nydf.csv')
df = pd.merge(df,nydf,on='Month')
```

<i>Now that the data is prepped, we can begin exploring it.  Here we can see simple average median rent prices per square foot in Newark, Hoboken, Jersey City, and New York City proper.</i>

```python
#highestrent = df.max(axis=0)
#lowestrent = df.min(axis=0)
#averagerent = df.mean(axis=0)

#exploring average rent data

averageNewarkNJ = "Newark, " + str(df.NewarkNJ.mean(axis=0))
averageHobokenNJ = "Hoboken, " + str(df.HobokenNJ.mean(axis=0))
averageJerseyCityNJ = "Jersey City, " + str(df.JerseyCityNJ.mean(axis=0))
averageShortHillsNJ = "ShortHills, " + str(df.ShortHillsNJ.mean(axis=0))
averageNewYorkNY = "New York, " + str(df.NewYorkNY.mean(axis=0))

averages = [averageNewarkNJ, averageHobokenNJ,averageJerseyCityNJ,averageNewYorkNY]

print averageNewarkNJ
print averageHobokenNJ
print averageJerseyCityNJ
print averageNewYorkNY
print averageShortHillsNJ
```

<i>The median prices are not all too surprising.  What does catch my attention is <b>Jersey City's relatively high median ZRI</b>.  Even taking into account areas like Journal Square and Newport where I imagine rent is very high, an overall median of \$1.99/sq.foot in Jersey City (compared to \$1.50/sq.foot in Short Hills - a relatively wealthy area - for example) piques my interest nonetheless. To put this into clearer context, we can visualize the data with <b>matplotlib</b>.</i>

<b>Objective</b>: Visualize New Jersey ZRI data with line graphs

```python
#graph settings

mpl.rc('lines', linewidth=5, color='r')
mpl.rc('font', size=9)

plt.figure(figsize=(25,5))
plt.xlabel('Month', fontsize=20)
plt.ylabel('$ per Sq. ft', fontsize=20)
plt.title('Median ZRI per sq. ft. ($)', fontsize=30, fontweight='bold')

df.HobokenNJ.plot(label='Hoboken, NJ', alpha=.7, xlim=(0,46))
df.JerseyCityNJ.plot(label='Jersey City, NJ', alpha=.7, xlim=(0,46))
df.BridgewaterNJ.plot(label='Bridgewater, NJ', alpha=.7, xlim=(0,46))
df.NewarkNJ.plot(label='Newark, NJ', alpha=.7, xlim=(0,46))
df.BedminsterNJ.plot(label='Bedminster, NJ', alpha=.7, xlim=(0,46))

legend = plt.legend(loc=2, shadow=True, fontsize=10)
legend.get_frame().set_facecolor('#eeeeee')
```

<b>Findings:</b>
* Since the beginning of 2011, rent prices have flucuated in <b>Hoboken, NJ</b> the most.  Not surprisingly, Hoboken has the highest rent levels.
* Prices in <b>Jersey City, NJ</b> have risen the most (and most steadily at that).  This could possibly be due to the gradual population growth in nearby Hoboken, NJ, and/or changes in nearby downtown Newark, NJ
* Rent levels in <b>Newark, NJ</b> seem to begin a steady upward trend in 2012, but overall are relatively stagnant - as are prices in <b>Bridgewater, NJ</b>, and <b>Bedminster, NJ</b> (these areas are included just for reference - mainly because I live near there)

<i>First a list is created containing all of the county seats.  Unfortunately, data for Freehold Borough, Hackensack, Mount Holly, and Salem weren't included in the set.  Instead, Freehold Township, Hasbrouck Heights, and Eastampton are substituted in.</i>

```python
#graph settings

mpl.rc('lines', linewidth=5, color='r')
mpl.rc('font', size=10)

seats1.plot(figsize=(25,5), xlim=(0,46))
seats2.plot(figsize=(25,5), xlim=(0,46))
seats3.plot(figsize=(25,5), xlim=(0,46))
```

<b>Objective</b>: Plot the averages of all New Jersey county seat median ZRI by sq. ft. in a bar graph

```python
means = (seats.mean())
plt.xlabel('County Seat', fontsize=20)
plt.ylabel('Average $ per Sq. ft', fontsize=20)
plt.title('Average Median ZRI - New Jersey County Seats', fontsize=30, fontweight='bold')
plt.colors()
means.plot(kind='bar', figsize=(25,5), color='#11ff88', alpha=0.3)
```

```python
maxs = (seats.max())
plt.xlabel('County Seat', fontsize=20)
plt.ylabel('Max $ per Sq. ft', fontsize=20)
plt.title('Maximum Median ZRI - New Jersey County Seats', fontsize=30, fontweight='bold')
plt.colors()
maxs.plot(kind='bar', figsize=(25,5), color='#11ff88', alpha=0.3)
```

#New Jersey Unemployment Rate vs. Median ZRI

<b>Exploration</b>: Bring in external data to test a variable against changes in New Jersey median ZRI levels.  For this run, state-level unemployment rates are used (BLS.gov).

```python
njue = pd.read_csv('https://raw.githubusercontent.com/c-trl/median-rent-prices-exploration/master/njue.csv')
njue = njue.set_index('Month')
njue.describe()
```

```python
njue.plot(kind='area', figsize=(25,5), color='#000000', alpha=0.3)
plt.title('New Jersey Unemployment Rates 2011 - 2014',fontsize=20,weight='bold')
plt.xlabel('Month', fontsize=15)
plt.ylabel('Unemployment Rate', fontsize=15)
plt.tick_params(axis='both', direction='out', length=6, width=2, colors='black', labelsize='large')

plt.show()
print njue[18:28]
```

<b>Here we see the declining trend in New Jersey's Unemployment Rate</b><sup>1</sup>, frome 2012-07 to 2013-04 when unemployment declined by 0.1% each month.  A preliminary hypothesis is that although it is illegal is some (to my knowledge, if not all) places to deny tenants who are unemployed, <b>I expect there to be at least minimal correlation between levels in renting and unemployment rate</b>.  Consequently, as demand for rentals may increase as unemployment decreases, it can be expected to find a slightly upward trend in median rent prices between 2012-07 and 2013-04.

Let's find out.

<sup>1</sup><small>[BLS.gov](http://data.bls.gov/timeseries/LASST340000000000003?data_tool=XGtable)</small>

```python
#graph settings
njue.plot(figsize=(15,5), kind='line', color='#111111')
nj.Hoboken.plot(figsize=(20,10))
nj.JerseyCity.plot(figsize=(20,10))

legend = plt.legend(loc=1, shadow=True, fontsize=12)
legend.get_frame().set_facecolor('#eeeeee')

plt.title('New Jersey Unemployment Rate (%) vs. Various Median ZRI ($/sq.ft)', fontsize=20, weight='bold')
plt.xlabel('Month', fontsize=15)
plt.ylabel('$/sq.ft', fontsize=15)
plt.twinx()
plt.xlim(0,46)
plt.ylim(0,10)

plt.show()
```

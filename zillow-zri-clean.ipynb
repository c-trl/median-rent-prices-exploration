#Importing CSV Files

```python
states = 'https://raw.githubusercontent.com/c-trl/zillow/master/state_zri.csv'
hoods = 'https://raw.githubusercontent.com/c-trl/zillow/master/neighborhood_zri.csv'
metros = 'https://raw.githubusercontent.com/c-trl/zillow/master/metro_zri.csv'
cities = 'https://raw.githubusercontent.com/c-trl/zillow/master/city_zri.csv'

dfs = pd.read_csv(states)
dfh = pd.read_csv(hoods)
dfm = pd.read_csv(metros)
dfc = pd.read_csv(cities)
```

```python
dfh.head()
```

###Cleaning Data, Rearranging Files

**State-Level Data**

```python
state_df = dfs.transpose()
state_df.columns=[x for x in state_df.iloc[0]]
state_df = state_df.drop('RegionName')

state_cols = state_df.columns.tolist()
state_cols = state_cols[-1:] + state_cols[:-1]
state_df = state_df[state_cols]
```

```python
state_df.head()
```

---

**Metro-Level Data**

```python
metro_df = dfm
metro_df = metro_df.transpose()
metro_df.columns = [x for x in metro_df.iloc[0]]
metro_df = metro_df.drop('RegionName')
```

```python
metro_df.head()
```

---

**City-Level Data**

```python
city_df = dfc
city_df['City'] = city_df.RegionName+', '+city_df.State
city_df = city_df.drop(['RegionName', 'State', 'Metro', 'CountyName'], axis=1)

city_cols = city_df.columns.tolist()
city_cols = city_cols[-1:] + city_cols[:-1]
city_df = city_df[city_cols]

city_df = city_df.transpose()
city_df.columns = [x for x in city_df.iloc[0]]
city_df = city_df.drop('City')
```

###Importing New Jersey Population Data

```python
url = 'http://www.usa.com/rank/new-jersey-state--population-density--city-rank.htm'
sourceCode = urllib2.urlopen(url).read()

topSplit = '<b>City / Population</b></td></tr>'
bottomSplit = '</table><p>Please'

sourceCodeSplit = sourceCode.split(topSplit)[1].split(bottomSplit)[0]
content = sourceCodeSplit.split('\n')
```

```python
cities = pd.DataFrame([re.findall(r'">(.*?)</a>', x) for x in content])
populations = pd.DataFrame([re.findall(r'</a> / (.*?)</td>', x) for x in content])
densities = pd.DataFrame([re.findall(r'</td><td>(.*?)/sq', x) for x in content])
```

```python
nj = pd.DataFrame()
nj['City'] = cities[0]
nj['Population'] = [str(x).replace(',', '') for x in populations[0]]
nj['Population Density'] = [str(x).replace(',', '') for x in densities[0]]

nj.head()
```

**Generating NJ Rent-level data:**

```python
nj_rent = city_df[[x for x in city_df.columns if 'NJ' in x]]

nj_rent = nj_rent.transpose().reset_index()
nj_rent['City'] = nj_rent['index']
nj_rent = nj_rent.drop(['index'], axis=1)

njr_cols = nj_rent.columns.tolist()
njr_cols = njr_cols[-1:] + njr_cols[:-1]
nj_rent = nj_rent[njr_cols]
```

```python
nj_df = nj.merge(nj_rent, on='City', how='left')

nj_df = nj_df.transpose()
nj_df.columns = [x for x in nj_df.iloc[0]]
nj_df = nj_df.drop('City')
nj_df = nj_df.dropna(axis=1)
```
